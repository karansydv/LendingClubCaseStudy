## Data cleaning and Manipulation

```python
# Remove the columns with Loan status as Current (as the problem statement is to analyse charged off and paid loan only)
df = df[~(df.loan_status=="Current")]
```

```python
df.shape
```

```python
# Following code drops the columns with more than half null values
df.dropna(axis= 1, how="all", thresh=df.shape[0]//2, inplace=True)
```

```python
df.shape
```

```python
# Remove all colums such as id, which have unique value
for i in df.columns:
    if(df[i].nunique() == df.shape[0]):  # if unique values in particular column is equal to number of rows
        print(i, df[i].nunique())        # print the columns with all unique values
        df.drop(i, axis=1, inplace=True)  # drop the column which have unique values in all rows
```

```python
# remonve the customer behavior variables are not available at the time of loan application, and thus they cannot be used as predictors for credit approval.
# variables such as acc_now_delinquent, chargeoff within 12 months etc. (which are related to the applicant's past loans) are available from the credit bureau.
```

#### It is seen that there is a mismatch of dataypes present i.e. few numerical columns are interpreted as object(str type) because of the presence of some special characters or wording in it. We have to make the data type as expected.

```python
# Term column has " months" after the numeric value
df.term = df.term.apply(lambda x: int(x.strip(" months")))  # code to strip the given trailing value
print(df.term.head())  
print(type(df.term))  # prints the type of the column
```

```python
# Int_rate column has "%" after the numeric value
df.int_rate = df.int_rate.apply(lambda x : float(x.strip("%")))   # code to strip the given trailing value
print(df.int_rate.head())
print(type(df.int_rate))  # prints the type of the column
```

```python
# Zip_code column has "xx" after the value
df.zip_code = df.zip_code.apply(lambda x : x[:-2])  # code to strip the given trailing value
print(df.zip_code.head()) # prints the type of the column
```

```python
# Displays the information about all the columns
df.info()
```

```python
# Fill the null values in emp_length column as "Other"
df.emp_length.fillna("Other", inplace=True)
```

- Creating bins to analyse categories the ranges present in annual_inc, int_rate, loan_amnt and dti

```python
loan_amnt_bins = [0,5000,10000,15000,20000,25000,40000]
loan_range = ['0-5000','5000-10000','10000-15000','15000-20000','20000-25000','25000 above']
df['loan_amnt_bins'] = pd.cut(df['loan_amnt'], loan_amnt_bins, labels=loan_range)
```

```python
annual_inc_bins = [0,25000,50000,75000,100000,1000000]
inc_range = ['0-25000','25000-50000','50000-75000','75000-100000','100000 above']
df['annual_inc_bins'] = pd.cut(df['annual_inc'], annual_inc_bins, labels=inc_range)
```

```python
int_rate_bins = [0,7.5,10,12.5,15,20]
int_rate_range = ['0-7.5', '7.5-10', '10-12.5', '12.5-15', '15 above']
df['int_rate_bins'] = pd.cut(df['int_rate'], int_rate_bins, labels=int_rate_range) 
```

```python
dti_bins = [0,5,15,20,25,30]
dti_range= ['0-5','5-10','10-20','20-25','25 above']
df['dti_bins'] = pd.cut(df['dti'], dti_bins, labels=dti_range, include_lowest=True)
```
